```python
df2 = pd.read_csv(path+"Kaggle/superheroes/super_hero_powers.csv")
df2.head(10)
```

## Description, nettoyage et jointure des datasets

```python
df1.rename(columns = {'name': 'Name'}, inplace = True)
df2.rename(columns = {'hero_names': 'Name'}, inplace = True)
df2.head()
```

```python
raw_data = pd.merge(df1, df2, on='Name')
raw_data.drop(['Unnamed: 0'], axis=1, inplace=True)
raw_data.head(10)
```

Ce croisement donne rien, il y a beaucoup trop de races ainsi que de Publisher. On va rationaliser les Publishers en comptant le nombre de heros qu'ils ont.

```python
raw_data['Weight'] = raw_data['Weight'].apply(lambda x: 0 if x <= 0
                                                    else 1 if x <= 25
                                                    else 2 if x <= 50
                                                    else 3 if x <= 100
                                                    else 4 if x <= 200
                                                    else 5)
```

## Script d'automatisation de groupby et de ratio

```python
# Exemple de generalisation de la fonction de ratio sur plusieurs colonnes d'un Dataframe

for dataset in raw_data.iloc[:, 7:10]: # On recupere les colonnes qui nous intéressent
    print(dataset.upper())
    
    for i in range (0,2): # On itere entre 0 et 1, 0 = Marvel and 1 = DC
        maxCatbyPublisher = len(raw_data.groupby(['Publisher',dataset]).count()['Name'][i]) # Calcul le max d'elements par Catégorie
        if i == 0:
            print (" -- Marvel Comics --")
        else :
            print ("-- DC Comics --")
        for j in range (1,maxCatbyPublisher): # On itere sur le max d'elements par Catégorie
            print(
                "ratio Categorie ", j, "/", j-1, " : ", 
                round(raw_data.groupby(['Publisher', dataset]).count()['Name'][i][j] 
                / raw_data.groupby(['Publisher', dataset]).count()['Name'][i][j-1], 2)) # Affichage du ratio
        
        print(" ")
```

## Est-ce que les superhéros féminins de Marvel et DC sont plutot good or evil ?

```python
# On va utiliser la syntaxe Query de Pandas qui est très utile mais moins rapide que d'autres
# Pour passer une varaible dans la Query, il faut mettre un @

for i in range (0,2): # On itere entre 0 et 1, 0 = Marvel and 1 = DC
    if i == 0:
        print (" -- Marvel Comics --")
    else :
        print ("-- DC Comics --")
    
    query_female = raw_data.query('Publisher == @i and Gender == "Female"')
    query_female_good =  query_female.query('Alignment == "good"')

    print(
            "Female : ", query_female['Name'].count(),
            "Female with good alignment : ", query_female_good['Name'].count(),
            "ratio good female", round((query_female_good['Name'].count() / query_female['Name'].count()),2)
    )
```

## On souhaite maintenant connaitre la proportion de Superhéros feminins chez Marvel et DC ainsi que le ratio de Superhéros masculins good pour relativiser la demande précédente ?

```python
# On va utiliser la syntaxe Query de Pandas qui est très utile mais moins rapide que d'autres
# Pour passer une varaible dans la Query, il faut mettre un @

for i in range (0,2): # On itere entre 0 et 1, 0 = Marvel and 1 = DC
    if i == 0:
        print (" -- Marvel Comics --")
    else :
        print ("-- DC Comics --")

    query_heros = raw_data.query('Publisher == @i')        
    query_female = raw_data.query('Publisher == @i and Gender == "Female"')
    query_female_good =  query_female.query('Alignment == "good"')
    query_male_good =  raw_data.query('Publisher == @i and Gender == "Male" and Alignment == "good"')


    print(
            "Heros ", query_heros['Name'].count(),
            "| Female ", query_female['Name'].count(),
            "| Male ", query_heros['Name'].count()-query_female['Name'].count(),
            "| ratio F ", round((query_female['Name'].count() / query_heros['Name'].count()), 2),
            "| Good F ", query_female_good['Name'].count(),
            "| ratio GF ", round((query_female_good['Name'].count() / query_female['Name'].count()),2),
            "| Good M ", query_male_good['Name'].count(),
            "| ratio GM ", round((query_male_good['Name'].count() / (query_heros['Name'].count()-query_female['Name'].count())),2)
    )
    
    plt.figure()
    plt.bar(["ratio Female", "ratio Good Female", "ratio Good Male"],
        [round((query_female['Name'].count() / query_heros['Name'].count()), 2), 
        round((query_female_good['Name'].count() / query_female['Name'].count()), 2),
        round((query_male_good['Name'].count() / (query_heros['Name'].count()-query_female['Name'].count())), 2)],
        color=['b','g','r']
    )
    plt.show()
```

On peut donc dire qu'il y autant de superhéros féminins chez Marvel que chez DC.
Que Marvel favorise plus que DC des superhéros feminins avec un alignement vers le Bien à plus de 80%.
Par contre que la parité de superhéros vers le Bien est respecté chez DC (environ 65% pour les deux sexes) alors que chez Marvel environ un superhéros masculin sur deux est aligné vers le Bien donc nettement moins que chez les femmes.

## Quels sont les super pouvoirs les plus et les moins répandus chez Marvel et DC ?

```python
# Le piege réside dans le fait qu'il peut y avoir plusieurs super pouvoris égaux à 1

score = [["none",1,"none"],["none",0,"none"]]

for i in range (0,2): # On itere entre 0 et 1, 0 = Marvel and 1 = DC
    if i == 0:
        print (" -- Marvel Comics --")
    else :
        print ("-- DC Comics --")
    
    score[1][1] = 0 #for second iteration
    q = raw_data.query('Publisher == @i')
    r = q.drop([
       'Publisher','Gender','Race','Height','Skin color','Eye color','Alignment','Hair color','Weight'], axis=1)

    for dataset in r:        
        count = int(r[dataset].value_counts().max())
        countmin = int(r[dataset].value_counts().min())

        if countmin == score[0][1]:
            print(dataset, countmin, " = ", score[0][1])
            score.append([dataset, countmin, i])

        elif count >= score[1][1]:
            print(dataset, count, " >= ", score[1][1])
            score[1][1] = count
            score.append([dataset, count,i])
            
print("---")            
print(score)
```

```python
for item in score:
    if item[0] == 'Agility':
        print(item)
```

```python
x = []
y = []

for item in score:  
    if item[2] == 0:
        x.append(item[0])
        y.append(item[1])
    
plt.figure(figsize=(18,6))
plt.bar(x,y)
plt.xticks(rotation = 90)
plt.xlabel('super powers')
plt.ylabel('# Marvel heroes')
plt.show()
```
