```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_path = "../../datasets/apple_store/AppleStore.csv"
dataset_description_path="../../datasets/apple_store/appleStore_description.csv"
```

```python
dataset = pd.read_csv(dataset_path, usecols=[2,3,5,8,12])
dataset.head()
```

```python
dataset['prime_genre'].value_counts()
```

```python
#data = dataset[dataset["prime_genre"] == 'Productivity']
data = dataset
data['name'] = data.loc[:,'track_name'].str.split(' ').str.get(0)
data = data.drop('track_name', axis=1)
```

```python
data['prime_genre']
```

```python
data['prime_genre'] = data['prime_genre'].apply(lambda x: 0 if x == 'Games'
                                                            else 1 if x == 'Entertainment' 
                                                            else 2 if x == 'Education' 
                                                            else 3 if x == 'Photo & Video' 
                                                            else 4 if x == 'Utilities' 
                                                            else 5 if x == 'Health & Fitness'  
                                                            else 6 if x == 'Productivity'
                                                            else 7 if x == 'Social Networking'
                                                            else 8 if x == 'Lifestyle'
                                                            else 9 if x == 'Music'
                                                            else 10 if x == 'Shopping'
                                                            else 11 if x == 'Sports'
                                                            else 12 if x == 'Book'
                                                            else 13 if x == 'Finance'
                                                            else 14 if x == 'Travel'
                                                            else 15 if x == 'News'
                                                            else 16 if x == 'Weather'
                                                            else 17 if x == 'Reference'
                                                            else 18 if x == 'Food & Drink'
                                                            else 19 if x == 'Business'
                                                            else 20 if x == 'Navigation'
                                                            else 21 if x == 'Medical'
                                                            else 22 if x == 'Catalogs'
                                                            else 23)
```

```python
data['prime_genre']
```

```python
data.sort_values(by = 'price', ascending=False)
```

```python
data.count()
```

```python
data['price'].value_counts(ascending=False)
```

```python
data['price'].plot(kind="hist")
plt.show()
```

```python
data.loc[data['price'] <= 0, 'price']                              = 0
data.loc[(data['price'] > 0) & (data['price'] <= 4.99), 'price']  = 1
data.loc[(data['price'] > 4.99) & (data['price'] <= 9.99), 'price']  = 2
data.loc[(data['price'] > 9.99) & (data['price'] <= 29.99), 'price']  = 3
data.loc[data['price'] > 29.99, 'price']                       = 4
```

```python
data['price'].value_counts()
```

```python
data['price'].plot(kind="hist")
plt.show()
```

```python
data['prime_genre'].plot(kind="hist")
plt.show()
```

```python
data.plot(kind="scatter",x='prime_genre',y='price')
plt.show()
```

```python
df = data.loc[:,('size_bytes', 'price')]
df['user-rating'] = data['user_rating'].astype(float)
df['prime-genre'] = data['prime_genre'].astype(float)
df
```

```python
import seaborn as sns
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Correlation Pearson des variables', y=1.05, size=15)
sns.heatmap(df.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```
